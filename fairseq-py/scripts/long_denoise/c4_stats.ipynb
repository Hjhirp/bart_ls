```python
# load length stats
from tqdm import tqdm
data_path = '/fsx/xwhan/data/pretrain_corpus/c4/bpe/stats0'
stats0 = []
tmp = 0
for line in tqdm(open(data_path).readlines()):
    wc = int(line.strip())
    if wc == 0:
        stats0.append(tmp)
        tmp = 0
    else:
        tmp += wc


data_path = '/fsx/xwhan/data/pretrain_corpus/c4/bpe/stats9'
stats9 = []
tmp = 0
for line in tqdm(open(data_path).readlines()):
    wc = int(line.strip())
    if wc == 0:
        stats9.append(tmp)
        tmp = 0
    else:
        tmp += wc
```

```python
import proplot as pplt
import numpy as np

data = stats9

fig, ax = pplt.subplots()
ax.hist(data, range=(0, 5000), bins=40, weights=np.ones(len(data)) / len(data))
```
